## 使用python连接neo4j，进行简单查询

```python
from neo4j import GraphDatabase

neo4j_config = {
    "NEO4J_URL": "bolt://localhost:7687", 
    "NEO4J_USER": "neo4j",
    "NEO4J_PASSWORD": "123456"
}

cypher_value = '''match p=(m)-[r]-(n) return p limit 1'''

driver = GraphDatabase.driver(neo4j_config['NEO4J_URL'], auth=(neo4j_config['NEO4J_USER'], neo4j_config["NEO4J_PASSWORD"]))

with driver.session() as session:
    return_result = session.run(cypher_value)
```

通过 可以看到 Result类有多个方法

比如 graph(), data(), value(), values()

通过  return_result?? 可以看到这些方法详细的返回信息

以 graph() 为例, 可以得到 cypher返回的图节点和关系及其所有类型、属性信息

```python
with driver.session() as session:
    return_result_graph = session.run(cypher_value).graph()
```

```python
for i,v in return_result_graph.__dict__.items():
    print(i,v, end='\n\n')
```
